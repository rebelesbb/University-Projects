```python
from importlib.metadata import distribution
from random import  choices, sample

# B | A "if A ? B" intrebam daca are loc b doar daca are loc intai a

urna = ['r'] * 5 + ['a'] * 3 + ['v'] * 2

def extragere() -> bool:
    bileExtrase = sample(urna, 3)
    if bileExtrase.count('r') >= 1:
        if bileExtrase.count(bileExtrase[0]) == 3:
            return True
        
    return False

def probabilitate():
    cnt = 0
    testCnt = 1000000
    for i in range(0, testCnt):
        if extragere():
            cnt += 1   
    print(cnt/testCnt)
            
    
            
probabilitate() 
        
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [randrange(1, 7) for _ in range(500)]
bin_edge = [k + 0.5 for k in range(0, 7)]
hist(data, bins=bin_edge, density=True, rwidth=0.9, color='green', edgecolor='black', alpha=0.5, label='frecvente_relative')

dist = dict([i, 1/6] for i in range(1, 7))
bar(dist.keys(), dist.values(), width=0.85, color='red', edgecolor='black', alpha = 0.6, label='probabilitati')
legend(loc = 'lower left')
grid()
show()
```
