```python
# rand int(365)
# facem de n ori, 1000 simulari
from random import randint
def birthday(n : int) -> bool:
    birthdayList = []
    isTheSame = False
    for i in range(n):
        bd =  randint(1, 365)
        if birthdayList.count(bd) > 0:
            isTheSame = True
            break
        birthdayList.append(bd)
    return isTheSame

def birthdayProb(n : int):
    count = 0
    for i in range(10000):
        if birthday(n):
            count += 1
    print(f'Probabilitatea ca doua persoane din {n} sa aiba aceeasi zi de nastere este : {count / 1000}')
    return count / 1000

birthdayProb(26)
        
```

```python
from matplotlib.pyplot import plot, grid, title, show, yticks, xticks
from numpy import arange
title('Plot test')
yticks(arange(0,1, step= 0.1))
xticks(range(0,60,5))
xs = range(2,50)
ys = [birthdayProb(x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
#1-A
```
